### Descripcion de el problema
#### La Programación Orientada a Objetos (POO) resuelve el desafío de modelado de la realidad mediante la abstracción, que consiste en convertir objetos físicos (como una planta) en estructuras lógicas (clases) para facilitar su gestión. El reto central es seleccionar solo los datos esenciales (atributos) y acciones específicas (funciones) que el sistema necesita, garantizando el encapsulamiento; esto significa que cada objeto es independiente y sus métodos actúan únicamente sobre su propia información a través del identificador self. En esencia, este enfoque permite organizar el código en unidades lógicas donde los datos y su comportamiento conviven, transformando la programación de una simple lista de pasos en un sistema estructurado de entidades que interactúan entre sí.

```python
class Planta:
    def __init__(self, nombreCientifico, tipo, fechaSiembra):
        # Atributos de la clase
        self.nombreCientifico = nombreCientifico
        self.tipo = tipo
        self.fechaSiembra = fechaSiembra

    # Funciones (métodos) de la clase
    def regar(self):
        return f"La planta {self.nombreCientifico} ha sido regada."

    def podar(self):
        return f"La planta {self.tipo} ha sido podada para un mejor crecimiento."

    def fertilizar(self):
        return f"Se ha aplicado fertilizante a la planta sembrada el {self.fechaSiembra}."

def crear_objeto_planta(numero):
    print(f"\n--- Ingresa datos para la planta {numero} ---")
    nombre = input("Nombre científico: ")
    tipo = input("Tipo (ej. Suculenta, Árbol): ")
    fecha = input("Fecha de siembra: ")
    return Planta(nombre, tipo, fecha)
# Crear objetos de forma interactiva
planta1 = crear_objeto_planta(1)
planta2 = crear_objeto_planta(2)

# Demostración de las funciones
print("\n--- Acciones ejecutadas ---")
print(planta1.regar())
print(planta1.podar())
print(planta2.fertilizar())
print(planta2.regar())

print("Programa Reealizado Por Keyla Paola Cigarroa Palacios NL =12")
```
